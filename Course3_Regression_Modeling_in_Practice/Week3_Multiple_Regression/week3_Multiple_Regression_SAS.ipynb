## Confidence Intervals
Confidence intervals tell us which values of the parameter estimates are plausible in the population. <br>
<br>
Typically, we look at 95% confidence intervals, which tell us, with 95% certainty, the range of parameter estimate values, that includes the true population parameter. That is, we are 95% certain that the true population parameter falls somewhere between the lower and upper confidence limits that are estimated based on our sample perimeter estimates. 

For example, the parameter estimate for H1ED9 (expelled from school) is 6.24. Meaning that on average, adolescents who are expelled from school have 6.24 more times skip school than adolescents who are not expelled from school. The conference interval for H1ED9 (expelled from school) is from 0.8 to 11.69. Meaning that 95% certain that the true population parameter for the assoication between expelled from school and times skip school fall somewhere between 0.8 to 11.69. That is, in the population there is a 95% chance that adolescents who are expelled from school have anywhere between 0.8 to 11.69 more times skip school than adolescents who are not expelled from school. <br>
<br>
Also note that "number of drinks have each time" variable H1TO16_C has a p-value of 0.072 which is not statistically significant, we cannot reject the null hypothesis or no association between expelled from school and number of drinks have each time after adjusted for expelled from school and health level in the model. The confidence interval for this variable is between -0.018 to 0.41, which includes a value of zero in that range, this means that we can't rule out with 95% confidence the probability that the association between "number of drinks have each time" and time skip school, after adjusting for other variables in the model, is zero. **In linear regression, when you have a nonsignificant P value, the 95% confidence interval for the parameter estimate will include a value of 0, no association. **

# Evaluating Model Fit

R-Square: the amount of variability in the response variable that is explained by our explanatory variable.

### Evidence of Misspecification
Specification is the process of developing a regression model. If a model is correctly specified, then the residuals, or error terms, are **not correlated with the explanatory variables**. If the data failed to meet the regression assumptions or if our model is missing important explanatory variables, then we have **model specification error**. 

### Regression Diagnostics
We perform regression diagnostics to try to understand the **cause of the misspecification** so that we can try to address it. 
1. DFFITS Statistic
2. Bartlett's Test
3. Kolmogorov-Smirnoff Test
4. Durbin-Watson Test
5. Variance Inflation Factor
6. **Model Residuals**

### Model Residuals
We can **assess violation of the assumptions** of the linear regression analysis by examining model residuals.

1. residual = res asks for a column with the unstandardized residuals and the option
2. student = res, asks for a column with the standardized residuals
3. out = results option asks for another data set that includes the regression analysis results
4. The information in the datasets produced by the output command allows us to generate additional regression diagnostic plots. 

The table shows that the coefficients for the linear and quadratic "number of drinks have each time" variable remain significant after adjusting for headache level, expelled from school and health level. <br>
Headache level is also statistically significant. The positive regression coefficient indicates that adolescents who are have an headache tend to have more times skip school. <br>
These explanatory variables together explain only about 7.3% (r-square=0.073) of the variability in time skip school.<br>

### Q-Q Plot

1. Q-Q Plot to evaluate the assumption that the residuals from our regression model are normally distributed. 
2. A Q-Q Plot plots the quantiles of the residuals that we would theoretically see if the residuals followed a normal distribution against the quantiles of the residuals estimated from our regression module. 

What we are looking for is to see if the points follow a **straight line**, meaning that the model estimated residuals are what we would expect if the residuals were normally distributed. As the Q-Q Plot shows, the residuals not follow a straight line, this indicates that our residuals do not follow perfect normal distribution. This could mean that the curvilinear association that we observed in our scatter plot may not be fully estimated by the quadratic "number of drinks have each time" term. There may be other explanatory variables that we might consider including in our model that could improve estimation of the observed curvilinearity.

### Leverage Plot
1. Identify observations that have unusually large influence on the estimation of the predicted value of the response variable
2. The leverage of an observation can be thought of in terms of how much the predict scores for other observations would differ if the observation in question were not included in the analysis.
3. The leverage always takes on values between zero and one. A point with zero leverage has no effect on the regression model. Outliers are observations with residuals greater than two or less than negative two. 

As the outliers are observations with residuals greater than two or less than negative two. The red symbols in the table are outliers, green symbols are observations which have high leverage values, brown symbols are observations which are outliers and have high leverage values. There are many outliers in the plot and they have small leverage values, meaning that although they are outlying observations, they do not appear to have a strong influence on the estimation of the regression parameters. On the other hand, we see that there are many cases with higher than average leverage, but they are not an outlier. There are 8 outliers in the plot.
