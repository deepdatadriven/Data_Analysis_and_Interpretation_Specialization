### Reference Group Coding / Reference Group Parameterization
when we exam explanatory variable group differences on response variable, reference group coding allows us to compare each group of our explanatory variable, referred to as the comparison groups, to another group, which is referred to as the reference group.

For example, if our response variable is times skip school, reference coding allows us to compare times skip school for each group of our categorical variable to a designated reference group. 


Headache level is a categorical explanatory variable which has five levels (0: never, 1: just a few times, 2: about once a week, 3: almost every day, 4: every day). I choose level 3 which is almost every day as my reference group, other levels are comparison groups on times skip school after controlling for the other explanatory variables in the model.

The reference group headache level 3 has a regression coefficient of 0, and no estimate of the standard error and p-value. The p-values of headache level 0, 1, 2, 4 are 0.031, 0.06, 0.1496, 0.2772. In addition, level 1, 2, 4 are significantly different from the level 3 in times skip school, because the p-values are exceed the alpha level of 0.05.
