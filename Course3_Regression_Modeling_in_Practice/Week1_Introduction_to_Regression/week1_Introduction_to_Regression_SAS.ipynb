# Determine whether the data is observational or experimental:
**Was the explanatory variable manipulated? or the values of the varable are controlled by the researcher?**
- **Yes, Manipulated**: experimental
- **No**: observational

***Statistically control*: include unbalanced covariates as additional explanatory varibles in the analysis.**<br>

***Observational data*: no researcher interaction or intervention with the explanatory variables**<br>
- **Data Reporting**: is the process of collecting and organizing data, typically to minitor a process or phenomenon. It often analyzed later on to test specific hypothesis. Data report can tell you what is happening, but data analysis can tell you why it's happening
- **Sample Survey**: observations is drawn from the population, and are asked to respond to questions posed by researcher, typically collected with certain hypothesis in mind to test.
- **Observational Study**: collected by simply observing the values of a variable or variables of interest as as they naturally occur.

***Experimental data*: observations are pre-selected or randomly assigned to values of the explanatory variable, and then observed on some response variable.**<br>
- **True Experiment**: <br>
1) Only one variable is manipulated, other variable are held constant<br>
2) Control group: no fertilizer=control group, fertilizer=treatment group<br>
3) Random assignment: check for imbalances between groups on covariates<br>
- **Quasi Experiment**： <br>
No random assignment<br>


# Confounding Variables
**In statistics, a confounding variable also known as confounding factor, a lurking variable, a confound or confounder is an extraneous variable (无关变量) that is associated positively or negatively with both the explanatory variable and response variable. **
1. Underlying weakness of most observational studies: besides of the certain value selected of observations, those observations may be different in additional ways that would also play a role in the response of interest
2. **Controlling for a third lurking variable** did not allow **a claim of causation**, only would be closer to establishing a causal connection
3. Observational studies cannot prove causality, observational studies are commonly used to draw conclusions about causal connections
4. Great care must be taken to control for the most likely lurking variables. Only then can researchers assert that observational study may suggest a causal relationship
