# Chi Square Test of Independent: Categorical -> Categorical
**Chi is a Greek letter that looks like a large X, so sometimes, you will see this test denoted with an X squared.**



**If the X^2 is not large than 3.84, it indicates that the data are not different enough from the null hypothesis for us to reject it.**

# SAS Program:

**categorical response variable * categorical explanatory variable / chisq;**

### **In my case, I will analyze if the SKIPCAT are associated with H1GH2**<br>
**Ho: There is no relationship between the two categorical variables(they are independent)**<br>
**Ha: There is a relationship between the two categorical variables(they are not independent)**<br>
**The Chi-Square statistic is calculated considering both the observed and expected counts in each of the table cells**<br>
/* Chi-square procedure */<br>
proc freq; tables SKIPCAT*H1GH2/chisq;<br>


Based on the information the above table provides: X^2 = 95.7340, 4 DF, p-value < .0001<br>
The p-value above of <.0001 clearly tells us that "times skipped school" and "headache level" are associated, which means that i will accept the alternate hypothesis that not all skip school rates are equal across headache level categories, but we don't know which are different and which are not since our explanatory variable has five categories.<br>
On the other hand, we can see the column percentage of SKIPCAT category 1, if we visualize the association, we can see that there seems to be a positive linear relationship, that is the more often adolescents have had a headache, the more likely times they skip school

### Post Hoc Test: Bonferroni Adjustment
If the categories of explanatory variable is more than two levels, we need to use post hoc test. For our case, we are using Bonferroni Adjustment method. The goal of using the Bonferroni Adjustment is to control a famliy-wise error rate, also known as the maximum overall type 1 error rate.<br>
1. **new p-value** = 0.05/(number of pairs)<br>
in our case: <br>
new p-value = 0.05/10 = 0.005<br>
2. run the chi-square test of independence for each paired comparison
data pair8; set new;<br>
if H1GH2=2 or H1GH2=3;<br>
proc sort; by aid;<br>
proc freq; tables SKIPCAT*H1GH2/chisq;<br>
run;<br>
