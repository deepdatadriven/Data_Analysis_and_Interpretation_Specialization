**A positive or increasing relationship, means that an increase in one of the variable is associated with an increase in the other<br>
A negative or decreaing relationship, means that an increase in one of the variable is associated with an decrease in the other.
PS: Not all relationships can be classified as either positive or negative.**

# Pearson Correlation Coefficient:
It measures a linear relationship between two quantitative variables


1. The numerical measure of a linear relationship between two quantitative variables is called **correlation coefficient** and is denoted by a lower cased **r**
2. The value of r range from **-1 to 1**
3. Negative values -1 - 0 of r indicate a negative direction for a linear relationship between the two variables.
4. Positive values 0 - 1 of r indicate a positive direction for a linear relationship between the two variables.
5. r values which are close to 0, whether they are negative or positive, indicate a week linear relationship.
6. r values which are close to -1 or +1 indicate a strong linear relationship, either negative or positive.

# Example: age -> distance



The scatter plot of the data shows a negative linear relationship between the age of a driver and the maximum distance at which a highway sign was legible.
The strength of the relationship revealed from the correlation r = -0.793. Since r is less than 0, it confirms that the direction of the relationship is negative. The r is relatively close to -1, it suggests that the relationship is moderately strong. So:<br>
**The negative correlation confirms that the maximum distance at which a sign is legible generally decreases with age.**<br>
**Higher age is associated with lower distance**<br>

**r^2**<br>
It is the fraction of the variability of one varibale that can be predicted by the other.

In my case, I analyze the relationship between "number of cigarettes smoked each day" and "numbers of drinks drunk each time":



I cannot get any information based on the scatter plot (Table 1), but based on the Pearson correlation coefficient r = 0.15860 and p-value <.0001(Table 2), it tells us the relationship is statistically significant. Since r is large than 0, it confirms that the direction of the relationship is positive. The r is relatively close to 0, it suggests that the relationship is weak. The positive correlation confirms that an increase in numbers of cigarettes is associated with an increase in numbers of drinks. On the other hand r^2 = 0.15860^2 = 0.002515, if we know the number of cigarettes, we can predict 0.25% of the variability we will see in the number of drinks. Of course, that also means that 99.75% the variability is unaccounted for.
