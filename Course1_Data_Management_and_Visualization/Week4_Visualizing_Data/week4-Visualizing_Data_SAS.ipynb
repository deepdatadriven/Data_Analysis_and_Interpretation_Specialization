# Describe distribution visually
**The overall pattern of the distribution of a quantitative variable, is described by its shape, center, and spread.**
1. choose intervals that correspond to the range of your variable
2. Draw a histogram graph based on the above intervals.
3. describe the overall pattern of the distribution, and mention any striking deviations from that pattern, based on the histogram's center, spread, and shape, while outliers will highlight deviations from that patten.
4. **Shape**: symmetry, skewness of a distribution, and peekness or modality, that is the number of peaks or modes that the distribution has.
<br>
1) Unimodal: it has one mode, roughly at ten, around which the observations are concentrated<br>
2) Bimodal: it has two modes, roughly at ten and tweenty, around which the observations are concentrated<br>
3) Flat/Uniform: the distribution which has no mode or no value around which the observations are concentrated. Instead the observations are roughly, uniformly distributed among the different values.<br>
5. mode: a value that occur most often
6. **Center**: a center of a distribution is its mid-point, the value that divides distribution to let approximately half the observations take smaller values, and approximately half take larger values.
7. **Spread/Variability**: can be described by the **approximate range** covered by the data

# Measures of center and spread
### The three main numerical measures of the center of the distribution:
1. Mean average<br>
the mean of n values = (sum of values) / n<br>
2. Modal average<br>
the mode = value with highest frequency(greater than 1)<br>
3. Median average<br>
the median = value of the middle item(or the mean of two middle items when arranged in order of magnitude)<br>
the scores are in ascending order and then look for the middle value.<br>

For example:<br>
scores: 2, 3, 3, 3, 12, 13<br>
mean = (2+3+3+3+12+13) / 6 = 6<br>
mode = 3<br>
median = (3+3)/2 = 3<br>

### Spread: Standard deviation
**Standard deviation**: is to quantify the spread of the distribution by measuring how far the observations are from their mean. Standard deviation gives the average or typical distance between a data point and the mean<br>
Variance (方差) = sum((observation value - mean)^2) / (number of observations - 1)<br>
Standard deviation (标准偏差) = sqrt(Variance) #squere root of this variance<br>
<br>
then we can know the average values will be in ***range/spread*** **mean ± standard deviation**<br>

# Appropriate descriptive statistics method for different format of variables
### Categorical variable:
Can be described by frequency tables: **proc freq; tables variable1, variable2;**
### Quantitative variable:
Can be described with histograms and exact measures of shape, center, and spread: **proc univariate; var variable1, variable2;**

```python
# categorical variable
proc freq; tables variable1, variable2;

# quantitative variable
proc univariate; var variable1, variable2;
```
