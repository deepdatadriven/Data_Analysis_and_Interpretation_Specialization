```python
import numpy as np
import pandas as pd

## load data
data = pd.read_csv("../data/addhealth.csv", low_memory=False)
## convert column names to upper case
data.columns = map(str.upper, data.columns)

## remove the limit and allows all rows and columns to be displayed
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
## only keep the variables that will be analysis
sub1 = data[['AID', 'H1GH2', 'H1DA1', 'H1DA2', 'H1DA4', 'H1DA5', 'H1DA6', 'H1DA7', 'H1ED1', 'H1ED2', 'H1DA10', 'H1GI1Y']]
sub2 = sub1[(sub1['H1GI1Y'] >= 80)]
sub = sub2.copy()
sub
```

```python
## convert object to numeric
sub['H1GH2'] = pd.to_numeric(sub['H1GH2'], errors='coerce')
sub['H1DA1'] = pd.to_numeric(sub['H1DA1'], errors='coerce')
sub['H1DA2'] = pd.to_numeric(sub['H1DA2'], errors='coerce')
sub['H1DA4'] = pd.to_numeric(sub['H1DA4'], errors='coerce')
sub['H1DA5'] = pd.to_numeric(sub['H1DA5'], errors='coerce')
sub['H1DA6'] = pd.to_numeric(sub['H1DA6'], errors='coerce')
sub['H1DA7'] = pd.to_numeric(sub['H1DA7'], errors='coerce')
sub['H1ED1'] = pd.to_numeric(sub['H1ED1'], errors='coerce')
sub['H1ED2'] = pd.to_numeric(sub['H1ED2'], errors='coerce')
sub['H1DA10'] = pd.to_numeric(sub['H1DA10'], errors='coerce')
```

```python
## coding out missing data
sub[['H1GH2', 'H1DA1', 'H1DA2', 'H1DA4', 'H1DA5', 'H1DA6', 'H1DA7']] = sub[['H1GH2', 'H1DA1', 'H1DA2', 'H1DA4', 'H1DA5', 'H1DA6', 'H1DA7']].replace([6,8], np.nan)
sub['H1ED1'] = sub['H1ED1'].replace([6, 7, 8, 9], np.nan)
sub['H1ED2'] = sub['H1ED2'].replace([996, 997, 998], np.nan)
sub['H1DA10'] = sub['H1DA10'].replace([996, 998], np.nan)
```

```python
sub.groupby('H1GH2').size()
```

```python
## create secondary variables
def times_skip(row):
    if row['H1ED1']==0:
        return 0
    elif row['H1ED1']==1:
        return 2
    elif row['H1ED1']==2:
        return 7
    elif row['H1ED1']==3:
        return 45
    else:
        return np.nan
    
sub['SKIPFREQ'] = sub.apply(lambda row: times_skip(row), axis=1)
```

```python
def total_skip(row):
    if row['SKIPFREQ'] != np.nan and row['H1ED2'] != np.nan:
        return row['SKIPFREQ']+row['H1ED2']
    elif row['SKIPFREQ'] == np.nan and row['H1ED2'] != np.nan:
        return row['H1ED2']
    elif row['SKIPFREQ'] != np.nan and row['H1ED2'] == np.nan:
        return row['SKIPFREQ']
    else:
        return np.nan
    
def skip_cate(row):
    if row['SKIPTOTAL']<7:
        return 0
    elif row['SKIPTOTAL']>=7:
        return 1
    else:
        return np.nan
    
sub['SKIPTOTAL'] = sub.apply(lambda row: total_skip(row), axis=1)
sub['SKIPCAT'] = sub.apply(lambda row: skip_cate(row), axis=1)
```

```python
sub['SKIPTOTAL'].describe()
```

```python
sub['SKIPCAT'].value_counts(dropna=False)
```

```python
def total_acti(row):
    return row['H1DA1']+row['H1DA2']+row['H1DA4']+row['H1DA5']+row['H1DA6']+row['H1DA7']

def acti_cate(row):
    if row['ACTIQU']<10:
        return 0
    elif row['ACTIQU']>=10:
        return 1
    else:
        return np.nan

sub['ACTIQU'] = sub.apply(lambda row: total_acti(row), axis=1)
sub['ACTICATE'] = sub.apply(lambda row: acti_cate(row), axis=1)
```

```python
sub['ACTIQU'].describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sub['SKIPCAT'] = sub['SKIPCAT'].astype('category')
## visualize categorical variable
ax=sns.countplot(x='SKIPCAT', data=sub)
plt.xlabel("skipped school more than 7 time or not")
plt.title("addhealth, times skipped scholl")
```

```python
## visualize quantitative variable, it will display the middle point of the intervals
ax1 = sns.distplot(sub['SKIPTOTAL'].dropna(), kde=False)
plt.xlabel("total times skipped school")
plt.title("addhealth, times skipped school")
```

```python
sub['SKIPBIN'] = pd.cut(sub.SKIPTOTAL, [0, 2, 6, 25, 110])
sub['SKIPBIN'].value_counts()
```
