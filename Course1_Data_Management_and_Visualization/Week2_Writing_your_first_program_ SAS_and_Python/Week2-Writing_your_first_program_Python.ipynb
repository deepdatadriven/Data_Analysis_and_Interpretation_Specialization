# Import data

```python
import pandas as pd
import numpy as np

data = pd.read_csv("../data/addhealth.csv", low_memory=False)
```

```python
print(len(data))
```

```python
print(len(data.columns))
```

```python
## change all column names to upper case
data.columns = map(str.upper, data.columns)
list(data.keys())
```

```python
# bug fix for display formats to avoid run time errors
pd.set_option("display.float_format", lambda x: '%f'%x)
```

```python
## convert values to numeric format
data['H1GH2'] = pd.to_numeric(data['H1GH2'], errors='coerce')
data['H1GH2'].head()
```

# Print counts and frequencies of a variable
## Use df.Series.value_counts(sort=False, normalize=True, dropna=False)

```python
c1 = data['H1GH2'].value_counts(sort=False, dropna=False)
c1
```

```python
p1 = data['H1GH2'].value_counts(sort=False, normalize=True)
p1
```

## Use data.groupby('H1GH2').size()

```python
ct1 = data.groupby('H1GH2').size()
ct1
```

```python
pt1 = data.groupby('H1GH2').size()*100/len(data)
pt1
```

# Create subset
## After sub dataset created, don't forget do a .copy() command

```python
data['H1GI3'] = pd.to_numeric(data['H1GI3'], errors='coerce')
sub1 = data[(data['H1GI3'] >= 3) & (data['H1GI3'] <= 18) & (data['H1ED2'] != 997)]
sub2 = sub1.copy()
len(sub2)
```
